```python
laliga=pd.read_csv("data/laliga.csv")
laliga
```

```python
laliga.groupby("Team")["Red Cards"].sum().nlargest(5)
```

```python
laliga.groupby("Position")["Long passes"].mean()
```

```python
laliga.groupby("Shirt number")["Goals scored"].sum().nlargest(10)
```

```python
shots = laliga.groupby("Team").agg(
    Total =("Shots","sum"), on_target=("Shots on target", "sum"))
shots
```

```python
shots["accuracy"]=shots["on_target"]/shots['Total']
shots["accuracy"].nlargest().plot(kind="barh")
shots["accuracy"].nsmallest().plot(kind="barh")
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2,1,figsize=(5,8), sharex=True)
shots["accuracy"].nlargest().sort_values(ascending = True).plot(kind="barh", ax=axs[0], color="Green", title="Most accurate Teams")
shots["accuracy"].nsmallest().sort_values(ascending = True).plot(kind="barh", ax=axs[1], color="Red", title= "Least accurate Teams")
```

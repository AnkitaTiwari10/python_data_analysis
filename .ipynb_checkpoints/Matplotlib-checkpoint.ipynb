```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nums = np.arange(5)
plt.plot(nums,nums, color="orange")
plt.plot(nums,nums**5, color="red")
plt.plot(nums,nums**7, color="black",linewidth = 4)
```

```python
nums = np.arange(5)
plt.plot(nums,nums, color="orange",linestyle = "dashed")
plt.plot(nums,nums**5, color="red",linestyle = "dashdot")
plt.plot(nums,nums**7, color="black",linewidth = 4, marker ="*", markersize = 25, markerfacecolor = "red")
```

```python
salary=[25000, 35000, 45000, 55000]
age = [20,30,40,50]
plt.plot(age,salary)
plt.xlabel("Age")
plt.ylabel("Salary")
```

```python
Fruits = ["Mango", "Orange", "Banana","Pineapple"]
Cost = [100,60,40,80]
plt.plot(Cost,Fruits)
plt.xlabel("Cost",labelpad=45, fontsize=20)
plt.ylabel("Fruits", labelpad=45,fontsize=20)
```

```python
plt.figure(figsize=(8,8))
plt.plot(nums,color = "teal", label="x",linestyle ="dotted")
plt.plot(nums*3,color = "red", label ="x cube",linestyle = "solid")
plt.plot(nums*9,color = "purple", label="x 9",linestyle="dashdot")
plt.legend()
```

```python
plt.figure(figsize=(8,8))
plt.plot(nums,color = "teal", label="x",linestyle ="dotted",)
plt.plot(nums*3,color = "red", label ="x cube",linestyle = "solid")
plt.plot(nums*9,color = "purple", label="x 9",linestyle="dashdot")
plt.legend(frameon=True, shadow=True,loc="lower right")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips= sns.load_dataset("tips")
```

# Scatter Plot


```python
tips
```

```python
penguins = sns.load_dataset("penguins")
```

```python
sns.scatterplot(data = tips, x = "total_bill", y= "tip", hue ="sex", style = "sex")
```

# Line Plot

```python
flights = sns.load_dataset("flights")
```

```python
sns.lineplot(data = flights, x="year", y="passengers", hue="month")
```

```python
sns.relplot(data = tips, x = "total_bill", y= "tip", kind="line", col = "sex", row="time")
```

```python
sns.relplot(data = tips, 
                x = "total_bill", 
                y= "tip",
                row="smoker",
                hue ="sex", 
                style = "sex")
```

# Histogram Plot

```python
sns.histplot(data = tips, x="tip")
```

```python
sns.histplot(data = tips, x="tip", hue="time")
```

```python
sns.histplot(data = penguins, x="body_mass_g", bins=20, hue="species")
```

```python
sns.histplot(data = penguins, x="body_mass_g", bins=20, hue="species", kde=True)
```

# Kde Plot

```python
sns.kdeplot(data= penguins, x="body_mass_g", hue="species", multiple="stack")
```

```python
sns.histplot(data = penguins, x="body_mass_g", y="flipper_length_mm")
```

```python
sns.kdeplot(data = penguins, x="body_mass_g", y="flipper_length_mm")
```

```python
sns.kdeplot(data = penguins, x="body_mass_g", y="flipper_length_mm", hue="species")
```

```python
sns.histplot(data = penguins, x="body_mass_g", y="flipper_length_mm", hue="species")
```

# Rug Plot

```python
sns.rugplot(data = penguins, x="body_mass_g", y="flipper_length_mm", hue="species")
```

```python
sns.kdeplot(data = tips, x="tip", hue="time")
sns.rugplot(data = tips, x="tip", hue="time")
```

```python
sns.violinplot(data = tips, x="tip", hue="time")
```

# Count Plot

```python
sns.countplot(data= penguins, x="species", hue="sex")
```

```python
Billboard[Billboard["rank"]==1]["artist"].value_counts().head(10).plot(kind="barh", edgecolor = "black", linewidth =3)
plt.title("Artists With Most Weeks #1")
plt.xlabel("Weeks At #1")
```

```python
Billboard[Billboard["song"]=="Please Mr. Postman"]["artist"].value_counts().plot(kind="pie", colors = ["#6c5ce7","#00cec9", "#74b9ff"], explode = [0,0,0.1])

plt.title("Please Mr. Postman Version")
```

```python
Billboard = pd.read_csv("data/billboard_charts.csv", parse_dates=["date"])

df = Billboard[Billboard["date"].between("2016-12-25","2021-01-01")]
a = df[df["song"]=="All I Want For Christmas Is You"]
a["rank"].plot(label="All I Want For Christmas Is You")
b =df[df["song"]=="Rockin' Around The Christmas Tree"]
b["rank"].plot(label="Rockin' Around The Christmas Tree")
c =df[df["song"]=="Jingle Bell Rock"]
c["rank"].plot(label="Jingle Bell Rock")
plt.title("Christmas Song Among The Top 100")
plt.ylabel("Chart Rank")
plt.xlabel("Date")
plt.legend()
```

```python
Billboard[Billboard["artist"]=="John Lennon"]["song"].value_counts("I'm Steppin' Out").plot(kind = "barh")

plt.figure(figsize=(14,8))
plt.subplot(2,2,1)
Billboard[Billboard["artist"]=="Wings"].value_counts("songs").head(8).plot(kind="barh")
plt.subplot(2,2,2)
Billboard[Billboard["artist"]=="George Harrison"].value_counts()["song"].value_counts("I'm Steppin' Out").head(8).plot(kind="barh")
```

```python
a= Billboard[Billboard["artist"]=="John Lennon"]["song"].value_counts(ascending = False).head(8)
b =Billboard[Billboard["artist"]=="Wings"]["song"].value_counts(ascending=False).head(8)
c =Billboard[Billboard["artist"]=="George Harrison"]["song"].value_counts(ascending=False).head(8)
d =Billboard[Billboard["artist"]=="Ringo Starr"]["song"].value_counts(ascending=False).head(8)
fig,axs = plt.subplots(2,2, figsize=(14,8))
plt.suptitle("Beatles Solo Songs (By Weeks On Charts", fontsize=20, )
a.plot(kind="barh", ax=axs[0][0], color="red")
b.plot(kind="barh", ax=axs[0][1], color="orange")
c.plot(kind="barh", ax=axs[1][0], color="blue")
d.plot(kind="barh", ax=axs[1][1],color="green")
plt.tight_layout()

```
